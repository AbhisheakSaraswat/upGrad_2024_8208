### Pandas Group by 

```python
df.head(2)
```

```python
df[["category","sales"]].groupby(by = 'category').sum()
```

```python
df[["category","sales"]].groupby(by = 'category').mean()
```

```python
df[["category","sales"]].groupby(by = 
                                 'category').agg({"sales" : ['sum','mean']})
```
