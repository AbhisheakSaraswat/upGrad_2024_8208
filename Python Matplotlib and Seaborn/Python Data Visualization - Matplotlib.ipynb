### Subplot function :

Using this particular function we can create subplot chart.

plt.subplot

* (1,2,1)
    * 1 Number of rows 
    * 2 Number of columns 
    * 1 The index of current subplot

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1) 
plt.hist(male['age'], bins = 20, color = 'blue')
plt.title("Histogram of age of Men")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.xticks(rotation=45)

plt.subplot(1,2,2) 
plt.hist(female['age'], bins = 20, color = 'green')
plt.title("Histogram of age of Women")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.yticks(range(0,35,5))
plt.suptitle("Histrograms of ages of Men and Women")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.barh(superstore.Category.value_counts().index,superstore.Category.value_counts().values)
plt.ylabel('count')
plt.xlabel('Category')
plt.title("Category wise count")
plt.grid(True, axis = 'y')
plt.show()
```

```python
superstore.head(2)
```

```python
result = pd.pivot_table(superstore, index='Sub-Category',
               values = ["Sales","Quantity","Profit"],
              aggfunc={"Sales":"mean","Quantity":"sum","Profit":"mean"}).reset_index()
```

```python
result.head()
```

```python
subcategory = result["Sub-Category"].to_list()
qty = result.Quantity.to_list()
sales = result.Sales.to_list()
profit = result.Profit.to_list()
```

```python
plt.figure(figsize=(8,3))
plt.bar(subcategory, qty, color = "c", label = "TotalQuantity")
plt.plot(subcategory, sales,marker = "x",linestyle="-",color="r", label="AvgSales")
plt.plot(subcategory, profit,'^--g', label="AvgProfit")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
payment_method = pd.crosstab(df_churn["PaperlessBilling"], 
                            df_churn["PaymentMethod"])
```

```python
payment_catogories = payment_method.columns
```

```python
payment_catogories
```

```python
x = np.arange(len(payment_catogories))
plt.bar(x, payment_method.loc['Yes'], label = "Yes")
plt.bar(x, payment_method.loc['No'], label = "No")
plt.xlabel("Payment methods")
plt.ylabel("Count")
plt.xticks(x, labels=payment_catogories)
plt.title("Distribution of number of payments over billing type and payment method")
plt.legend()
plt.show()
```
