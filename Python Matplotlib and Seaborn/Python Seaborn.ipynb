### Line Plot

```python
plt.figure(figsize=(6,4))
sns.lineplot(data = amazon, x = "Date", y = 'Open', label = "Opening Price")
sns.lineplot(data = amazon, x = "Date", y = 'Close', label = "Closing Price")
plt.xlabel("Date")
plt.ylabel("Opening price in USD($)")
plt.title("Opening price of Amazon Stock")
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data = df, x = "TotalCharges")
plt.title("Distribution of total charges")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data = df, x = "TotalCharges", bins = 30, color = "linen")
plt.title("Distribution of total charges")
plt.show()
```

### Stacked Histogram

```python
plt.figure(figsize=(5,3))
sns.histplot(data = df, x = "TotalCharges", hue = "PaymentMethod",
             multiple='stack')
plt.title("Distribution of total charges split by payment method")
plt.xlabel("Total Charges")
plt.show()
```

### Bar Chart

```python
df.head()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = df, x = "PaymentMethod")
plt.title("Number of customer per payement method")
plt.xticks(rotation=90)
plt.xlabel("Payment method")
plt.ylabel("Cnt")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = df, y = "PaymentMethod")
plt.title("Number of customer per payement method")
plt.xticks(rotation=90)
plt.xlabel("Count")
plt.ylabel("Payment method")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = df, y = "PaymentMethod", hue = "PaperlessBilling")
plt.title("Number of customer per payement method")
plt.xticks(rotation=90)
plt.xlabel("Count")
plt.ylabel("Payment method")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.barplot(data = df, y = "PaymentMethod", x = "TotalCharges", estimator="mean", errorbar=None)
plt.title("Number of customer per payement method")
plt.xticks(rotation=90)
plt.xlabel("Count")
plt.ylabel("Payment method")
plt.show()
```

```python
df.head(1)
```

```python
plt.figure(figsize=(10,3))
bar_plot = sns.catplot(data = df, kind = 'bar', x = "TotalCharges", y = "PaymentMethod",
                      col = "PaperlessBilling", estimator='mean', errorbar=None)
bar_plot.set_axis_labels('Average Total charges', 'Payment method')
bar_plot.fig.suptitle("Avg total charges per payment method split by billing type", y = 1.02)
plt.show()
```
