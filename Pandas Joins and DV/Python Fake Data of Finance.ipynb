```python
import pandas as pd
from faker import Faker
import random

# Initialize the Faker instance
fake = Faker()

# Set the number of rows you want to generate
num_rows = 1000

# Generate a dataset
data = {
    "Company": [fake.company() for _ in range(num_rows)],
    "Stock Symbol": [fake.lexify(text='????') for _ in range(num_rows)],  # Stock symbols with random 4 letters
    "Transaction Date": [fake.date_between(start_date='-2y', end_date='today') for _ in range(num_rows)],
    "Transaction Type": [random.choice(['Buy', 'Sell', 'Dividend', 'Interest']) for _ in range(num_rows)],
    "Shares": [random.randint(1, 1000) for _ in range(num_rows)],
    "Share Price ($)": [round(random.uniform(5, 500), 2) for _ in range(num_rows)],
    "Transaction Amount ($)": [round(random.uniform(1000, 100000), 2) for _ in range(num_rows)],
}

# Convert to a pandas DataFrame
finance_df = pd.DataFrame(data)

# Save to a CSV file
finance_df.to_csv('synthetic_finance_data.csv', index=False)

# Display the first few rows of the dataset
finance_df.head()
```
