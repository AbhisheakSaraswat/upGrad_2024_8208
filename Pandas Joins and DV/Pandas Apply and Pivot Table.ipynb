```python
import pandas as pd
import numpy as np
import warnings 
import os
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Superstore.csv", encoding="ISO-8859-1")
```

```python
df.head(2)
```

```python
df = df[["Order Date","Ship Mode","Customer Name","Region",
         "Segment","State","Category",
  "Sub-Category","Sales","Quantity","Profit"]].copy()
```

```python
df.head(2)
```

```python
# user defined function
def calc_sales_bucket(sale):
    if sale >= 0 and sale <= 500:
        return "Sales 0 - 500"
    elif sale > 500 and sale <= 1000:
        return "Sales 500 - 1000"
    elif sale > 1000 and sale <= 5000:
        return 'Sales 1000 - 5000'
    else:
        return "Sales > 5000"
```

```python
df["Bucket"] = df.apply(lambda x: calc_sales_bucket(x['Sales']), axis = 1)
```

```python
df.head()
```

```python
df["Bucket"].unique()
```

```python
df["Bucket"].value_counts().reset_index()
```

```python
df.head()
```

```python
df["Qty_2"] = df["Quantity"].apply(lambda x : x ** 2)
```

```python
df.head()
```

```python
def sales_profit(row):
    return row["Sales"] - row["Profit"]
```

```python
df["AcutalSales"] = df.apply(sales_profit, axis = 1)
```

```python
df.head()
```

```python
df['lower_cust'] = df['Customer Name'].apply(lambda x : x.lower())
```

```python
df.head()
```

```python
df.drop(columns=["Bucket","Qty_2","AcutalSales","lower_cust"], inplace = True)
```

```python
df.head(2)
```

```python
df["LenCust"] = df["Customer Name"].apply(len)
```

```python
df.to_csv("Modified_Superstore.csv", index = False)
```

### Pivot Table 

In Python, Pandas Pivot table calc average of numbers column by default.

```python
pd.pivot? # data, columns, index, values
```

```python
pd.pivot_table? 
# data, columns, index, values 
# aggfunc, fill_value, margins, dropna, margins_names,observed, sort
```

```python
df.head()
```

```python
pd.pivot_table(data = df, values = "Sales", index = "State").head()
```

```python
pd.pivot_table(data = df, values = "Sales", index = "State",
              columns="Segment", fill_value = "0")
```

```python
pd.pivot_table(data = df, values = "Sales", index = "Sub-Category",
              columns="Segment", aggfunc=sum, fill_value = "0")
```

```python
pd.pivot_table(data = df, values = "Sales", index = "Sub-Category",
              columns="Segment", 
               aggfunc=[sum,"mean",'max'], fill_value = "0")
```

```python
pd.pivot_table(data = df, values = ["Sales","Quantity"],
               index = ["Region","Category"],
              columns="Segment",
              aggfunc={"Sales" : "mean", "Quantity" : [sum,'mean']})
               
```

```python
result = pd.pivot_table(data = df, values = ["Quantity"],
               index = ["Category"],
              columns="Segment",
              aggfunc= sum, margins=True, margins_name="GrandTotal")
               
```
