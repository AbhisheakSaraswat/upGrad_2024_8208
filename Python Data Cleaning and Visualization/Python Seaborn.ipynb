```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(data = tips, x = "total_bill", y = 'tip', palette='husl',
               hue = 'sex', style = "time")
plt.show()
```

```python
sns.relplot(data = tips, x = "total_bill", y = 'tip',hue = 'sex', style = "time")
plt.show()
```

```python
gap = px.data.gapminder()
```

```python
gap.head()
```

```python
gap_ind = gap[gap['country'] == "India"]
```

```python
gap_ind.head()
```

```python
plt.figure(figsize = (2,3))
sns.relplot(data = gap_ind, x = "year", y = "lifeExp", kind = 'line')
plt.show()
```

```python
ipj = gap[gap['country'].isin(["India","Pakistan","Japan"])]
```

```python
ipj.head()
```

```python
plt.figure(figsize = (2,2))
sns.relplot(data = ipj, x = "year", y = "lifeExp", hue = 'country',
            kind = 'line')
plt.show()
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index = "country",
                                                 columns = 'year',
                                                 values = 'lifeExp')
```

```python
temp_df.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot = True, linewidths=0.5)
plt.title("YoY Life Expectancy")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.barplot(data = tips, x = 'day', y = 'total_bill', ci = True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = tips, x = 'day',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = tips, x = 'day',hue='sex', palette=['brown','grey'])
plt.show()
```

### Regression Plot:

```python
plt.figure(figsize=(5,3))
sns.regplot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```
