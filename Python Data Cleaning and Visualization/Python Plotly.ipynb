```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_stocks = px.data.stocks()
```

```python
df_stocks.head()
```

```python
px.line(df_stocks, x = 'date', y = 'GOOG', template='gridon',height=300, width=700)
```

```python
px.line(df_stocks, x = 'date', y = 'GOOG', template='plotly_dark',height=300, width=700)
```

```python
px.line(df_stocks, x = 'date', y = ['GOOG','AAPL'], title = 'Apple vs Google',
       height=400, width=750, template="gridon")
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.AAPL, mode = 'lines', name = "Apple"))
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.AMZN, mode = 'lines+markers',
                         name = "Amazon"))
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.GOOG, mode = 'lines+markers',
                         name = "Google", line = dict(color = "firebrick", width=3,
                                                      dash='dashdot')))
fig.update_layout(title = "Stock Pricde Data 2018-2020",
                 xaxis_title="Date", yaxis_title="Price", template="plotly_dark")
```

```python
df = pd.read_csv("Superstore.csv", encoding="ISO-8859-1")
```

```python
df.head(2)
```

```python
df.info()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df.dtypes
```

```python
result = df.groupby(by = df["Order Date"])["Sales"].sum().reset_index()
```

```python
result.head(2)
```

```python
px.line(result, x = "Order Date", y = "Sales", title = 'TotalSales',
       height = 500, width = 1200, template="plotly_dark")
```

```python
df['MonthName'] = df["Order Date"].dt.month_name()
```

```python
df.head(2)
```

```python
result = df.groupby(by = 'MonthName')["Sales"].sum().reset_index().sort_values(by="Sales",
                                                                              ascending=False)
```

```python
result.head()
```

```python
px.bar(result, x = "MonthName", y = "Sales", title="Month wise Sales", color = "MonthName",
      labels = {"Sales" : "Total Sales"}, height=400, width=780, template="seaborn")
```

```python
result = df[["Order Date","Region", "State","City","Category","Sales","Quantity"]]
result.head(2)
```

```python
stateSales = result.groupby(by="State")["Sales"].sum().reset_index().sort_values("Sales",
                                       ascending=False).reset_index(drop=True)
```

```python
stateSales.head()
```

```python
px.treemap(stateSales, path = ["State"], values = "Sales", height=400,width=800,template="gridon")
```

```python
get_sales = result.groupby(by = ["State","City"])["Sales"].sum().reset_index().sort_values(
    "Sales",ascending=False).reset_index(drop=True)
```

```python
get_sales.head()
```

```python
px.treemap(get_sales, path = ["State","City"],
           values = "Sales", height=600,width=800,template="ggplot2")
```

```python
region_sales = result.groupby(
    by = ["Region","State",
          "City"])["Sales"].sum().reset_index().sort_values(
    "Sales",ascending=False).reset_index(drop=True)
```

```python
region_sales.head()
```

```python
px.treemap(region_sales, path = ["Region","State","City"],
           values = "Sales", height=600,width=800,template="ggplot2")
```
