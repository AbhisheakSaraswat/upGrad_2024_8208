Using this package we can send the HTTP request like Get, Post, Put, Delete

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.geeksforgeeks.org/python-programming-language-tutorial/"
response = requests.get(url)
print("Status Code: ", response.status_code)
```

```python
print("URL: ", response.url)
```

```python
print(response.content)
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
# title of the page
```

```python
soup.title
```

```python
soup.title.text
```

### HTML Tags

* div - To wrap sections of page and apply styles or scripts to them.
* title - Specifies the title of the webpage
* p (paragraphs)
* h1 - h6 we have headings
* body - contains the main content of HTML document.
* header
* footer
* img
* li (list of individual items)
* ol (Creates a numbered list
* ul (Creates a bulleted list)

```python
s = soup.find('div', class_="wrapper single-page")
p = s.find_all('p')
lstOfParagraphs = []
for i in p:
    lstOfParagraphs.append(i.text)
    
```

```python
print(lstOfParagraphs)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(lstOfParagraphs, columns=['Paragraphs'])
```

```python
df.to_csv("Paragraphs.csv", index = None)
```

```python
headings = s.find_all(['h1','h2','h3','h4','h5','h6'])
```

```python
headings
```

```python
main = soup.find("div", id="table_of_content")
data = main.find_all("li")
for i in data:
    if len(i.text) > 0:
        print(i.text)
```

```python
links = soup.find_all('a')
for link in links:
    href = link.get("href")
    if href:
        print(href)
```

```python
paragraphs = soup.find_all('p')
for p in paragraphs:
    print(p.text)
    print()
```

```python
images = soup.find_all('img')
for img in images:
    image_url = img.get('src')
    print(image_url)
```
