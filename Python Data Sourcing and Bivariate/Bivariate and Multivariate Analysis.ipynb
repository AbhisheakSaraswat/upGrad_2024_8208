### BiVariate:

The main objective of bivariate analysis is to understand how one variable changes in relation to another.

### Key Types of Bivariate Analysis:
* For Categorical Data Set
    * Chi-Square Test 
    * Cross-tablulation - Frequency distribution of 2 categorical variables in matrix format.

* For Numerical Data
    * Scatter Plots
    * Correlation Analysis
    
* For Categorical vs Numerical Data
    * T-Test or ANOVA
    
    
### Use of Bivariate Analysis
* Indentifying Relationship
* Hypothesis Testing
* Predictive Analysis
* Decision Making

### Multivariate Analysis:

When we want to check the relationship between three or more variables simultaneously.

### Key Types of Multivariate Analysis:
* Multiple Linear Regression
* Logistic Regression
* Cluster Analysis
* PCA 

### Use of It
* Complex Relationship
* Predicive Modeling
* Quality Control
* Market Research
* Publich Health

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("titanic_dataset.csv")
```

```python
df.head(2)
```

### Bivariate Analysis

Scatter Plot:

Relationship between two continuous (numerical) values.

```python
plt.figure(figsize=(5,3))
sns.boxplot(x = df['Pclass'], y = df['Fare'])
plt.title("Box Plot of Fare vs Pclass")
plt.show()
```

```python
df.columns
```

```python
categorical_cols = ["Survived","Pclass",'Sex','SibSp','Parch','Embarked']
continuous_cols = ['Fare',"Age"]
```

```python
for cat_col in categorical_cols:
    for continuous in continuous_cols:
        print(cat_col, 'vs', continuous)
        plt.figure(figsize=(5,3))
        sns.boxplot(x = df[cat_col], y = df[continuous])
        plt.title(f"Boxplot: {cat_col} vs {continuous}")
        plt.show()
    
```

### barplot:

It is used to see statistical distribution (mean/median/var/std dev/max/min etc) of a continuous variable against a categorical variable.

```python
plt.figure(figsize=(5,3))
sns.barplot(x = df['Pclass'], y = df['Fare'], ci = None)
```

```python
lstOfcombination = []
for cat_col in categorical_cols:
    for continuous in continuous_cols:
        lstOfcombination.append((cat_col,continuous))
        print(cat_col, 'vs', continuous)
        plt.figure(figsize=(5,3))
        sns.barplot(x = df[cat_col], y = df[continuous])
        plt.show()
    
```

```python
# plt.figure(figsize=(3,2))
pd.crosstab(df['Sex'], df['Survived']).plot(kind = 'bar', stacked = True)
plt.title('Stacked Bar Plot - Sex vs Survived')
plt.ylabel('Count')
plt.show()
```

### Pairplot:
It is used to see pairwise relationships between two continuous numerical variables in the dataset.

```python
plt.figure(figsize=(5,3))
sns.pairplot(df[['Age','Fare']].dropna())
plt.show()
```

```python
# Countplot showing the number of passengers by Pclass and survied
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x = 'Pclass', hue = 'Survived', data = df)
plt.title("Count Plot - Passengers by Pclass and Survived")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.kdeplot(data = df, x = 'Age', hue = 'Survived', fill = True)
plt.title("KDE Plot - Age with Survived as hue")
plt.show()
```
